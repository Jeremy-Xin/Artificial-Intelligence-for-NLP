```python
filename = '80k_articles.txt'
```

```python
all_content = open(filename).read()
```

```python
all_content[:200]
```

```python
import re

def tokenize(string):
    return ''.join(re.findall('[\w\d]+', string))
```

```python
all_content = all_content.replace('\\n','')
```

```python
all_content = tokenize(all_content)
```

```python
all_content[:200]
```

# Unigram

```python
from collections import Counter
```

```python
char_counter = Counter(all_content)
```

```python
char_counter.most_common()
```

```python
from matplotlib.pyplot import yscale, xscale, title, plot

M = char_counter.most_common()[0][1]

yscale('log')
xscale('log')
title('Frequency of n-th most frequent word and 1/n line.')
plot([c for (w,c) in char_counter.most_common()])
plot([M/i for i in range(1, len(char_counter))])
```

```python
def get_prob_from_counter(counter):
    all_occurences = sum(counter.values())
    def get_prob(item):
        return counter[item] / all_occurences
    return get_prob

get_char_prob = get_prob_from_counter(char_counter)
print(char_counter)
```

```python
from functools import reduce
from operator import mul
```

```python
def prob_of_string(string):
    return reduce(mul, [get_char_prob(c) for c in string])
```

```python
prob_of_string('这是一个比较罕见的测试用例')
```

```python
prob_of_string('这是一个比较常见的测试用例')
```

```python
print(prob_of_string('广州有一个地方叫做沥窖'))
print(prob_of_string('杭州有一个地方叫做西湖'))
```

```python
pair1 = '前天晚上吃晚饭的时候', '前天晚上吃早饭的时候'
pair2 = '正是一个好看的小猫', '真是一个好看的小猫'
pair3 = '我无言以对，简直', '我简直无言以对'

pairs = [pair1, pair2, pair3]
```

```python
def get_probability_performance(language_model_func, pairs):
    for p1, p2 in pairs:
        p1, p2 = tokenize(p1), tokenize(p2)
        print('*' * 20)
        print('\t {} with probability {}'.format(p1, language_model_func(p1)))
        print('\t {} with probability {}'.format(p2, language_model_func(p2)))
```

```python
get_probability_performance(prob_of_string, pairs)
```

# 2-Gram

```python
all_content[:100]
```

```python
gram_length = 2
two_gram_counter = Counter(all_content[i:i + gram_length] for i in range(len(all_content) - gram_length))
```

```python
two_gram_counter.most_common()[:20]
```

```python
get_pair_prob = get_prob_from_counter(two_gram_counter)
```

```python
def get_2_gram_prob(word, prev):
    if get_pair_prob(prev + word) > 0:
        return get_pair_prob(prev + word) / get_char_prob(prev)
    else:
        return get_char_prob(word)

def get_2_gram_string_prob(string):
    probabilities = []
    for i, c in enumerate(string):
        prev = '<>' if i == 0 else string[i - 1]
        probabilities.append(get_2_gram_prob(c, prev))
    return reduce(mul, probabilities)
```

```python
get_probability_performance(get_2_gram_string_prob, pairs)
```
