```python
import numpy as np
import pandas as pd
from subprocess import check_output
import jieba
print(check_output(["ls", "."]).decode("utf8"))
```

```python
def process(input_path, data_type):
    data = pd.read_csv(input_path)
    x = data[['content']]
    y = data.drop(['id','content'], axis=1)
    
    dummies = []
    for c in y.columns:
        dummy = pd.get_dummies(y[c], prefix=c)
        dummies.append(dummy)
    y_labels = pd.concat(dummies, axis=1)
    
    def cut(sentence):
        return ' '.join(jieba.cut(sentence))
    x['content'] = x['content'].map(cut)
    
    x.to_csv('./processed/{}_x_processed.csv'.format(data_type), index=False)
    y_labels.to_csv('./processed/{}_y_processed.csv'.format(data_type), index=False)
```

```python
process('./input/sentiment_analysis_trainingset.csv', 'train')
process('./input/sentiment_analysis_validationset.csv', 'val')
process('./input/sentiment_analysis_testa.csv.csv', 'test')
```
